```python
# ===============================================
# üß† Building a Text Classification Model using
# Bag-of-Words (BoW) + SVM (Support Vector Machine)
# ===============================================

# Step 1Ô∏è‚É£: Import CountVectorizer to convert text ‚Üí numeric features
from sklearn.feature_extraction.text import CountVectorizer

# Step 2Ô∏è‚É£: Create a simple class to store category names
class Category:
  BOOKS = "BOOKS"
  FASHION = "FASHION"

# Step 3Ô∏è‚É£: Training data (sentences + their category)
# Each sentence belongs to either BOOKS or FASHION category
sentences = [
    "I love the books",
    "This is a great book",
    "Reading books makes me happy",
    "Book pages are of high quality",
    "I love the shoes",
    "The fit is great",
    "These shoes are stylish",
    "Fashion trends change every year"
]

# Step 4Ô∏è‚É£: Corresponding category labels
categories = [
    Category.BOOKS,
    Category.BOOKS,
    Category.BOOKS,
    Category.BOOKS,
    Category.FASHION,
    Category.FASHION,
    Category.FASHION,
    Category.FASHION
]

# Step 5Ô∏è‚É£: Convert text sentences into numeric vectors (Bag of Words model)
vectorizer = CountVectorizer()
sentences_vectors = vectorizer.fit_transform(sentences)

# To see the vocabulary and feature matrix
# print(vectorizer.get_feature_names_out())   # Shows all unique words
# print(sentences_vectors.toarray())          # Shows numeric representation

# building a classifier to classify whether the sentence is a book category or clothing category

# Step 6Ô∏è‚É£: Import SVM model
from sklearn import svm

# Step 7Ô∏è‚É£: Create an SVM classifier with linear kernel
# (Linear kernel works well for text data)
clf_svm = svm.SVC(kernel="linear")

# Step 8Ô∏è‚É£: Train the model on the training data
clf_svm.fit(sentences_vectors, categories)

# Step 9Ô∏è‚É£: Test the model with a new, unseen sentence
testing_classifier = vectorizer.transform(["The Quality of the book pages are so good"])

# Step üîü: Predict the category (BOOKS or FASHION)
prediction = clf_svm.predict(testing_classifier)

# Step 1Ô∏è‚É£1Ô∏è‚É£: Print the result
print("Predicted Category:", prediction[0])
```
