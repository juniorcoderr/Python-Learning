```python
from sklearn.feature_extraction.text import CountVectorizer

sentences = [
    "I love programming",
    "Programming is fun",
    "I love Python"
]

# creating object
cv = CountVectorizer()

# Convert text to numbers
X = cv.fit_transform(sentences)

# to check vocabulary
print(cv.get_feature_names_out())

# to print the matrix
print(X.toarray())
```
